```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import konlpy
```

```python
import json
with open('10003.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
 
```

```python
from konlpy.tag import Okt
from konlpy.tag import Kkma

okt = Okt()
kkma = Kkma()
```

```python
texts = []
for item in data:
    text = item['review'] 

    texts.append(text)
    
print(texts)
```

```python
print(okt.morphs(text))
```

```python
text = str(texts)

print(okt.pos(text))
```

```python
print(kkma.pos(text))
```
